 - False Positive (FP): The model predicts that a customer will churn, but the customer does not actually churn.  
   Impact :  
   * Company lost customer acquistion cost
   
 - False Negative (FN): The model predicts that a customer will not churn, but the customer actually does churn.
   Impact:
   * Company lost repuchase revenue
   * Comoany lost retaining cost

 Since acquisition and repurchase revenue have high impact on business performance, we provide model that combine FP and FN cost to provide a more accurate picture of the overall performance of the model with F1 metric score.

#### Categorical Feature vs Target
To examine the relationship between categorical features and the target variable, we will use the chi-square test. The hypotheses for this test are as follows:
- Null Hypothesis (Ho): The categorical feature and the target Churn are independent.
- Alternative Hypothesis (Ha): The categorical feature and the target Churn are dependent.


The chi-square test is used because it is well-suited for assessing associations between categorical variables by comparing the observed frequencies in each category with the frequencies expected if there were no association. This allows us to determine whether any significant dependency exists between the categorical features and the target variable.

```python
# Perform Chi-Square Test
categorical_features = list(df.describe(include='O').columns)
chi2_pvalues = []

for feature in categorical_features:
    chi_table = pd.crosstab(df[feature], df['Churn'])
    stat, p_value, dof, expected = chi2_contingency(chi_table)
    chi2_pvalues.append(round(p_value, 5))

chi2_summary = pd.DataFrame({
    'categorical_feature': categorical_features,
    'p_value': chi2_pvalues
})
chi2_summary['notes'] = np.where(chi2_summary['p_value'] > 0.05, 'independent', 'dependent')
chi2_summary
```

Based on the chi-square test results above, all categorical features have a chi-square score below 0.05, which leads us to reject the null hypothesis. This indicates that each categorical feature is significantly associated with the target variable `Churn`, meaning that the proportions differ significantly

### Handling Outliers

```python
num_features = df.select_dtypes(include=[np.number])

def find_outliers(data):
    q1 = data.quantile(0.25)
    q3 = data.quantile(0.75)
    iqr = q3 - q1
    limit = iqr * 1.5
    lower_limit = q1 - limit
    upper_limit = q3 + limit
    print(f'IQR: {iqr}')
    print(f'Lower bound: {lower_limit}')
    print(f'Upper bound: {upper_limit}')

for feature in num_features.columns:
    print(f'\nIQR of {feature}:')
    find_outliers(num_features[feature])
```

### Do customers who churn on the e-commerce subscription tend to stop using the service during the initial months of usage?

```python
# Function to create a crosstab table visualization
def create_crosstab(var1, var2):
    ct_table = pd.crosstab(df[var1], df[var2])
    ct_table.loc['Total'] = ct_table.sum()
    ct_table.loc[f'{var1} %'] = round(ct_table.loc[1] / ct_table.loc['Total'] * 100, 1)
    return ct_table

plt.figure(figsize=(16,6))

# Create a histogram for 'Tenure' with 'Churn' as hue
hist_ax = sns.histplot(data=df, x='Tenure', hue='Churn')
hist_ax.set_title('Tenure by Actived vs Churned Customer', size=14, weight='bold')
hist_ax.set_xlabel(None)
plt.legend(title='Customer Status', labels=['Churned', 'Actived'])

# Display the transposed crosstab table for 'Churn' vs 'Tenure'
create_crosstab('Churn', 'Tenure').T.head(5)
```

Histograms show that although e-commerce usage is high during the first two months, over 50% of customers churn during that period (early-life churn). This indicates that even though customers are initially attracted, many quickly leave the service.

According to [Cartesian's article](https://www.cartesian.com/monitoring-and-mitigating-early-life-churn/), early churn is often triggered by:
- **Customer Support:** Issues with complaint resolution.
- **Product Availability:** Products not consistently available.
- **Service Quality:** Suboptimal payment and delivery processes.

***Recommendations***
- **Service Optimization:** Improve customer support, ensure product availability, and streamline transaction processes.
- **Gamification Implementation:** Introduce a points or leveling system to enhance customer engagement and retention from the start.

These steps can help reduce early-life churn and increase customer loyalty.

Some factors that may influence the phenomenon where customers with high satisfaction scores actually show a higher churn rate include:

- **High Expectations:** Customers who give high satisfaction scores may have very high expectations for the service. When the service fails to meet these ideal expectations, they tend to quickly seek alternatives.
- **Fragile Loyalty:** Even if satisfied, customers with high expectations are often more critical and easily inclined to switch to competitors if they find a more attractive offer.
- **Gap Between Experience and Expectation:** A discrepancy between the actual experience and the ideal expectation can lead to sudden disappointment, even though customers initially feel satisfied.

Research by Anderson and Fornell (1994) suggests that high customer satisfaction does not always translate into long-term loyalty, especially when external factors such as better competitor offerings are present. Additionally, Keiningham et al. **References:**

- Keiningham, T. L., Aksoy, L., Cooil, B., & Andreassen, T. W. (2007). "Linking Customer Satisfaction to Customer Loyalty, Retention, and Profitability." *Journal of Service Research*, 9(2), 121-133.

### Does the effectiveness of promotions and discounts influence churn?

```python
# Boxplot for CouponUsed based on Churn status
plt.figure(figsize=(8,6))
sns.boxplot(data=df, x='Churn', y='CouponUsed', palette="Set2")
plt.title("Distribution of Coupon Usage Based on Churn Status")
plt.xlabel("Churn (0 = Non-Churn, 1 = Churn)")
plt.ylabel("Coupon Used")
plt.show()

# Displaying summary statistics from the boxplot using print
# Using groupby to get descriptive statistics for each Churn group
print("Summary statistics for CouponUsed based on Churn status:")
print(df.groupby('Churn')['CouponUsed'].describe())

# Alternatively, to display the specific values used in the boxplot,
# we calculate the quartiles and whiskers for each group.

print("\nBoxplot values (quartiles and whiskers) for CouponUsed:")

# Grouping data based on Churn status
for churn_value, group in df.groupby('Churn'):
    # Calculating quartiles
    Q1 = group['CouponUsed'].quantile(0.25)
    median = group['CouponUsed'].median()
    Q3 = group['CouponUsed'].quantile(0.75)
    IQR = Q3 - Q1
    # Lower whisker: the minimum value that is within Q1 - 1.5*IQR
    lower_whisker = group['CouponUsed'][group['CouponUsed'] >= (Q1 - 1.5 * IQR)].min()
    # Upper whisker: the maximum value that is within Q3 + 1.5*IQR
    upper_whisker = group['CouponUsed'][group['CouponUsed'] <= (Q3 + 1.5 * IQR)].max()
    
    print(f"\nChurn Status = {churn_value} ({'Non-Churn' if churn_value == 0 else 'Churn'}):")
    print(f"  Q1 (25th percentile)   = {Q1:.2f}")
    print(f"  Median (50th percentile) = {median:.2f}")
    print(f"  Q3 (75th percentile)     = {Q3:.2f}")
    print(f"  Lower Whisker          = {lower_whisker:.2f}")
    print(f"  Upper Whisker          = {upper_whisker:.2f}")
```

Based on the summary statistics and boxplot values presented, several analytical points can be drawn:

**Same Core Distribution:**
- **Quartiles and Median:** For both groups (Churn = 0 and Churn = 1), the values of Q1, median, and Q3 are identical (Q1 = 1, median = 1, Q3 = 2). This indicates that 50% of the central data (values between Q1 and Q3) are the same for both groups, suggesting no fundamental difference in coupon usage levels among the majority of customers.

**Consistent Whiskers:**
- **Whisker Boundaries:** For both non-churn and churn groups, the lower whisker is 0 and the upper whisker is 3. This shows that the range of non-extreme values is the same between the two groups.
- **Outliers:** Although the maximum value reaches 16 in both groups, these outliers do not affect the boxplot since the whiskers stop at 3. This indicates that most customers have low coupon usage values (primarily within the range of 0–3).

**Almost the Same Mean:**
- **Mean:** The average coupon usage is very similar between non-churn (mean ≈ 1.77) and churn (mean ≈ 1.79) groups, with only a negligible difference likely influenced by outliers that are not represented in the boxplot.

**Implications for Promotion Effectiveness:**
- **Preliminary Conclusion:** Since the coupon usage distribution is nearly identical in both groups—in terms of quartiles, median, and whisker boundaries—it can be concluded that the variable CouponUsed (serving as a proxy for promotion and discount effectiveness) does not show a significant difference between customers who churn and those who do not.
- **Interpretation:** In other words, the level of coupon usage does not appear to differ between customers who remain subscribed and those who churn. This suggests that, based on the current data, coupon-based promotions alone do not significantly impact the churn decision.

Overall, the visualization and summary statistics indicate that coupon usage among customers is fairly homogeneous between the two groups. Thus, coupon promotions by themselves are likely insufficient to distinguish between customers who will churn and those who will not. Further analysis incorporating other promotional variables (such as CashbackAmount or additional promotional parameters) may be necessary to gain a more comprehensive understanding.


```python
plt.figure(figsize=(8,6))
sns.boxplot(data=df, x='Churn', y='CashbackAmount', palette="Set3")
plt.title("Distribution of Cashback Amount Based on Churn Status")
plt.xlabel("Churn (0 = Non-Churn, 1 = Churn)")
plt.ylabel("Cashback Amount")
plt.show()


# Displaying Summary Statistics for CashbackAmount

print("Summary statistics for CashbackAmount based on Churn status:")
print(df.groupby('Churn')['CashbackAmount'].describe())

# Calculating specific quartile and whisker values for each group
print("\nBoxplot values (quartiles and whiskers) for CashbackAmount:")
for churn_value, group in df.groupby('Churn'):
    Q1 = group['CashbackAmount'].quantile(0.25)
    median = group['CashbackAmount'].median()
    Q3 = group['CashbackAmount'].quantile(0.75)
    IQR = Q3 - Q1
    lower_whisker = group['CashbackAmount'][group['CashbackAmount'] >= (Q1 - 1.5 * IQR)].min()
    upper_whisker = group['CashbackAmount'][group['CashbackAmount'] <= (Q3 + 1.5 * IQR)].max()
    
    label = "Non-Churn" if churn_value == 0 else "Churn"
    print(f"\nChurn Status = {churn_value} ({label}):")
    print(f"  Q1 (25th percentile)   = {Q1:.2f}")
    print(f"  Median (50th percentile) = {median:.2f}")
    print(f"  Q3 (75th percentile)     = {Q3:.2f}")
    print(f"  Lower Whisker          = {lower_whisker:.2f}")
    print(f"  Upper Whisker          = {upper_whisker:.2f}")


# Statistical Test: t-test for CashbackAmount

group0_cashback = df[df['Churn'] == 0]['CashbackAmount']
group1_cashback = df[df['Churn'] == 1]['CashbackAmount']

t_stat_cashback, p_val_cashback = stats.ttest_ind(group0_cashback, group1_cashback, equal_var=False)
print("\nResults of the t-test for CashbackAmount:")
print("t-statistic =", t_stat_cashback)
print("p-value =", p_val_cashback)
```

Based on the statistical results and t-test for **CashbackAmount**, here are the insights:

- **Churn (Churn = 1):** The average cashback is **161.79**.

The higher upper whisker for the Non-Churn group suggests that most customers in this group receive higher cashback values, even though both groups have similar outliers with the same maximum values.

**t-test Results:**
